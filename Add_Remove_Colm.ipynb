### Adding/Removing Columns

```python
import pandas as pd
```

```python
coffee = pd.read_csv('./Data/coffee.csv')
```

```python
coffee['Price'] = 4.99
```

```python
import numpy as np

coffee['new_price'] = np.where(coffee['Coffee Type']=='Espresso', 3.99, 5.99)
```

```python
coffee.drop(columns=['Price'], inplace=True)
```

```python
coffee.tail()
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price'] # multiply units to price in new revenue column
```

```python
coffee.rename(columns={'new_price':'Price'}, inplace= True)
```

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else 'Average' if x < 185 else 'Tall') # using lambda function add 'height_category'   
```

```python
# using function to categorize athlete according to height and weight
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'

    else:
        return 'Heavyweight'


bios['Category'] = bios.apply(categorize_athlete, axis=1)
```
